```python
import os #To access local folders
import json
import pandas as pd
import mysql.connector
```

```python
# Aggregated Insurance

agg_ins_folder_path = "D:/Guvi/Projects/Phonepe Pulse Data Visualization and Exploration/Data/pulse/data/aggregated/insurance/country/india/state/"
folder_list = os.listdir(agg_ins_folder_path)
#print(folder_list)

agg_ins_columns = {"State":[],"Year":[],"Quarter":[],"Type":[],"Count":[],"Amount":[]}

for state in folder_list:
    state_path=agg_ins_folder_path+state+"/"
    state_list=os.listdir(state_path)
    #print(state_list)
    
    for year in state_list:
        year_path=state_path+year+"/"
        year_list=os.listdir(year_path)
        #print(year_list)

        for file in year_list:
            file_path=year_path+file
            with open(file_path,"r+") as json_data:
                data_agg_ins = json.load(json_data)
                #print(data_agg_ins)

                try:
                    for i in data_agg_ins["data"]["transactionData"]:
                        name=i["name"]
                        count=i["paymentInstruments"][0]["count"]
                        amount=i["paymentInstruments"][0]["amount"]
                        #print(name)

                        agg_ins_columns["State"].append(state)
                        agg_ins_columns["Year"].append(year)
                        agg_ins_columns["Quarter"].append(int(file.strip(".json")))

                        agg_ins_columns["Type"].append(name)
                        agg_ins_columns["Count"].append(count)
                        agg_ins_columns["Amount"].append(amount)
                        #print(agg_ins_columns)
                except:
                    pass

df_agg_insurance = pd.DataFrame(agg_ins_columns)
```

```python
# Table Creation

connection= mysql.connector.connect(host="localhost",user="root",password="SQL@2023",database="phonepe_data")
mycursor=connection.cursor()

# Aggregated Insurance Table
query1 = '''create table if not exists agg_ins(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              Type varchar(255),
                                              Count bigint,
                                              Amount float)'''
mycursor.execute(query1)

# Aggregated Transaction Table
query2 = '''create table if not exists agg_trans(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              Type varchar(255),
                                              Count bigint,
                                              Amount float)'''
mycursor.execute(query2)

# Aggregated User Table
query3 = '''create table if not exists agg_user(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              Brand varchar(255),
                                              Count bigint,
                                              Percentage float)'''
mycursor.execute(query3)

# Map Insurance Table
query4 = '''create table if not exists map_ins(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              District varchar(255),
                                              Count bigint,
                                              Amount float)'''
mycursor.execute(query4)

# Map Transaction Table
query5 = '''create table if not exists map_trans(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              District varchar(255),
                                              Count bigint,
                                              Amount float)'''
mycursor.execute(query5)

# Map User Table
query6 = '''create table if not exists map_user(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              District varchar(255),
                                              RegisteredUsers bigint,
                                              AppOpens bigint)'''
mycursor.execute(query6)

# Top Insurance Table
query7 = '''create table if not exists top_ins(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              Pincode bigint,
                                              Count bigint,
                                              Amount float)'''
mycursor.execute(query7)

# Top Transaction Table
query8 = '''create table if not exists top_trans(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              Pincode bigint,
                                              Count bigint,
                                              Amount float)'''
mycursor.execute(query8)

# Top User Table
query9 = '''create table if not exists top_user(State varchar(255),
                                              Year bigint, 
                                              Quarter int, 
                                              Pincode bigint,
                                              RegisteredUsers bigint)'''
mycursor.execute(query9)
```
